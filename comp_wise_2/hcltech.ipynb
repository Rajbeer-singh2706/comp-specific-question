```python




SELECT 
    id,
    CASE 
        WHEN SUM(CASE WHEN country = 'Global' THEN 1 ELSE 0 END) > 0 
              THEN MAX(CASE WHEN country = 'Global' THEN value END) 
        ELSE SUM(value)
    END AS total
FROM temp
GROUP BY id;

1     [ uk, global, ,ge, ine 
ID  = [ 0 , 10 , 1, 0 , 0 ]




group by ---> 1
```
